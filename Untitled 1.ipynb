```python
import pandas as pd
```

```python
import pandasql as ps
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('concessions_adjusted.csv')
```

```python
df['Revenue']=df['Revenue'].str.replace('$','')

df['Revenue']=df['Revenue'].str.replace('(','')

df['Revenue']=df['Revenue'].str.replace(')','').astype(float)
```

```python
a=df.groupby('Stand Name').agg({ 'Revenue': 'sum'})
```

```python
# Top 15 and bottom hotels and their reve
b=a.sort_values(by=['Revenue'],ascending=False)

import matplotlib.pyplot as plt

%matplotlib inline
plt.figure(figsize=(20,10))
plt.bar(b.index[:15],b['Revenue'][:15],width=0.8, align='center')
plt.xticks(rotation=45)
plt.ylabel('Revenue in $')
plt.title('Top 15 Highest Generating Revenue Stands ')
plt.savefig('hotel_revenue_top15.png')
plt.show()
```

```python
b=a.sort_values(by=['Revenue'],ascending=True)

import matplotlib.pyplot as plt

%matplotlib inline
plt.figure(figsize=(20,10))
plt.bar(b.index[:15],b['Revenue'][:15],width=0.8, align='center')
plt.xticks(rotation=45)
plt.ylabel('Revenue in $')
plt.title('Bottom 15 Stands and their Revenue')
plt.savefig('hotel_revenue_b15.png')
plt.show()
```

Time Series Analysis

```python
a=df.groupby('Stand Name').agg({ 'Quantity': 'sum'})
a=a.sort_values(by=['Quantity'],ascending=False)

d=a[:15]
```

```python
df['Order Time']=df['Order Create Time'].astype(str)+':00'

df['Order Time']=pd.to_datetime(df['Order Time'])
df=df.set_index('Order Time')
df.index = pd.to_datetime(df.index,infer_datetime_format=True)

a=df.groupby('Stand Name').resample('15min',base=30, label='right').agg({ 'Revenue': 'sum'}).reset_index()
a = a[a['Stand Name'].isin(d.index)]
```

```python
p=df.groupby('Category Group').resample('15min',base=30, label='right').agg({ 'Quantity': 'sum'}).reset_index()
```

```python
p.pivot(index='Order Time',columns="Category Group", values="Quantity").plot(figsize=(8,4),title='quantity v/s time for each product category')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
a.pivot(index='Order Time',columns="Stand Name", values="Revenue").plot(figsize=(15,7),title='revenue in $ v/s time for top 15 Restaurant')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
